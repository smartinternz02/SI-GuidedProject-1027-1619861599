```python
import seaborn as sn
```

```python
iris = sn.load_dataset('iris')
```

```python
iris
```

```python
sn.scatterplot(x = "sepal_length",y = "petal_length",data=iris)
```

```python
sn.scatterplot(x = "sepal_width",y = "petal_width",data=iris)
```

```python
sn.lineplot(x = "sepal_length",y = "petal_length",data=iris)
```

```python
sn.lineplot(x = "sepal_width",y = "petal_width",data=iris)
```

```python
sn.pairplot(iris)
```

```python
sn.barplot(x="species",y="petal_length",data=iris)
```

```python
sn.barplot(x="species",y="sepal_length",data=iris)
```

```python
sn.barplot(x="species",y="petal_width",data=iris)
```

```python
sn.barplot(x="species",y="sepal_width",data=iris)
```

From above 4 barplots we can say that length of petal,sepal and width of sepal is more in virginica.In setosa sepal width is more.

```python
sn.boxplot(x="sepal_length",y="species",data=iris)
```

```python
sn.heatmap(iris.corr(),annot=True)
```
